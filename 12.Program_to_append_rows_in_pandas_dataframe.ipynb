# Program to append rows in pandas dataframe using for loop

```python
import pandas as pd
import numpy as np
df=pd.read_csv('ipl.csv')
display(df)
print()
a=list(df)
b=[]
values=pd.read_csv('ipl1.csv',header=None)
y=values.shape[0]
x=0
for i in range(y):
    d=list(values.iloc[x])
    c=zip(a,d)
    c_dict=dict(c)
    b.append(c_dict)
    x+=1
b1=pd.DataFrame(b)
display(b1)
df2=df.append(b,ignore_index=True)
display(df2)
```
