# Program to handle missing data in pandas datframe

```python
import pandas as pd
movie=pd.read_csv('movie.csv')
display(movie)
values={"Actor":"Siddharth Malhotra","Actress":"No Actress","Movies":"Badhaai Do"}
movie.fillna(value=values,inplace=True)
movie['Release_Year']=pd.to_datetime(movie['Release_Year'])
y=round(movie['Budget in crores'].median())
movie['Budget in crores'].fillna(y,inplace=True)
x=round(movie['Theatres'].mean())
movie['Theatres'].fillna(x,inplace=True)
movie.dropna(inplace=True)
movie
```
