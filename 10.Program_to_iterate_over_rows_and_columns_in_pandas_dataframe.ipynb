# Program to iterate over rows and columns in pandas dataframe

```python
import pandas as pd
df=pd.read_csv("data1.csv")
display(df)
print("\n Iterrows Function")
for index,row in df.iterrows():
    print(index,row.Calories<400)
print("\n Itertuples Function\n")
for row in df.itertuples():
    
    print(row.Pulse>110 & row.Pulse<140)
print("\nIteritems Function\n")
for key,value in df.iteritems():
    print('Key : ',key)
    print('Values : ',value.values)
```
