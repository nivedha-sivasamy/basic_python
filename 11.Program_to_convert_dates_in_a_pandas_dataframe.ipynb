# Program to convert dates in a pandas dataframe to a date time data type

```python
import pandas as pd
data=pd.read_csv("Cricket.csv")
display(data)
print()
data['DOB']=pd.to_datetime(data['DOB'])
display(data)
print()
data['DOB']=data['DOB'].dt.strftime("%B %d %Y")
display(data)
```
